```python
import numpy as np
import matplotlib as plt
import control as ct

from math import sqrt

param = {
    'A': 1.00,
    'beta': 0.75,
    'a12': 0.05,
    'a0': 0.02,
    'g': 9.81
}

def update(t, x, u, param):
    # Obter as entradas
    v, w = u

    # Obter os estados
    h1, h2 = x

    # Obter os parâmetros
    A = param.get('A')
    beta = param.get('beta')
    a12 = param.get('a12')
    a0 = param.get('a0')
    g = param.get('g', 9.81)

    # Definir as atualizações dos estados
    alpha = a12 * (1. / A) * np.sqrt(2. * g) * np.sqrt(h1 - h2)
    h1_dot = beta * (1. / A) * v - alpha + w
    h2_dot = alpha * a0 * (1. / A) * np.sqrt(2. * g) * np.sqrt(h2)

    return np.array([h1_dot, h2_dot])

def output(t, x, u, param):
    # Obter os estados
    h1, h2 = x

    # Obter a saída
    y = h2

    return np.array([y])

system = ct.nlsys(update, output, inputs=('v', 'w'), states=('h1', 'h2'), outputs='h2')
print(system)
```

```python
import numpy as np
import matplotlib as plt
import control as ct

from math import sqrt

# Definição de matrizes

A = np.array([[-1., -1.],
              [6.5, 0.]])

B = np.array([[1., 1.],
              [1., 0.]])

C = np.array([[1., 1.],
              [1., 0.]])

D = np.array([[0., 0.],
              [0., 0.]])

system = ct.ss(A,B,C,D)

# Definição de série temporal

final_time = 10.
step = 0.01
timepts = np.arange(0, final_time, step)

# Definição de entrada
U1 = np.heaviside(timepts, 1)
U2 = np.heaviside(timepts, 1)

# Definição de simulador
response = ct.input_output_response(sys=system, U=[U1, U2], T=timepts)

response.plot()
```
